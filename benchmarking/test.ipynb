```python
n = 100
p = 10
df = (
    pl.DataFrame({'patient_id': list(range(n))})
        .with_columns(
            treatment_id=pl.col('patient_id').map_elements(lambda x: np.arange(p)[np.random.binomial(n=1, p=0.2, size=p).astype(bool)].astype(str)),
        )
        .filter(pl.col('treatment_id').list.len() > 0)
        .with_columns(
            reward=pl.col('treatment_id').map_elements(lambda ids: np.random.standard_exponential(len(ids))),
            cost=pl.col('treatment_id').map_elements(lambda ids: np.random.standard_exponential(len(ids))),
        )
)

# That's the data generation part

# Need the connector for maq package
```

```python
from sparse_maq import Solver

unique_patient_ids = df.select('patient_id')
unique_treatment_ids = df.select('treatment_id').explode('treatment_id').unique().select(treatment_id=pl.col('treatment_id'))
solver = Solver(unique_patient_ids, unique_treatment_ids)
```

```python
solver.fit(df)
```
